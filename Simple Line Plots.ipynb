```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

# remember using as = plt.axes() is the more obejct orientated way to go
```

```python
# first step, make a figure
fig = plt.figure()
```

```python
# then the axes
ax = plt.axes()
```

```python
# put things on the axes
x = np.linspace(0, 10, 1000)
ax = plt.axes()
ax.plot(x, np.sin(x))
# ax.plot(x, np.cos(x), color='red')
# ax.plot(x, np.cos(x), linestyle='dotted')
# ax.plot(x, np.cos(x), linestyle='--')
ax.plot(x, np.cos(x), '--r') # dashed cyan
```

```python
x = np.linspace(0, 10)
ax = plt.axes()
ax.set_xlim(-1, 11)
ax.set_ylim(-1.5, 0.98);
ax.plot(x, np.cos(x), ':r')
```

```python
# set the axes things all at once
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(
    xlim=(0, 10),
    ylim=(-2, 2),
    xlabel='x',
    ylabel='sin(x)',
    title='A Simple Plot',
)
```

```python
# scatter plots
ax = plt.axes()
x = np.linspace(0,10,100)
ax.scatter(x, np.cos(x),
           c=['red' if b else 'blue' for b in np.cos(x)>0],
           alpha=0.7
          )
```

```python
# ERROR BARS
ax = plt.axes()
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy #* np.random.randn(50)

ax.errorbar(x, y, yerr=2, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
# contour plots
fig, ax = plt.subplots()
# ax = plt.axes()
x = np.linspace(0,10,100)
X, Y = np.meshgrid(x,x)
Z = np.cos(X) + np.sin(Y)
ax.contour(X, Y, Z, 40, cmap='RdGy')
```

```python
# histogram
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(-0.5, 1.2, 1000)
ax = plt.axes()
ax.hist(x1, bins=30, normed=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='black')
ax.hist(x2, bins=30, normed=True, alpha=0.5, histtype='stepfilled',
         color='orange', edgecolor='green')
```
