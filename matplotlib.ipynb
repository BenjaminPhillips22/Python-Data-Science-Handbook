```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# set the style
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
# save image
fig.savefig('my_figure.png')
```

```python
# There are two main ways to plot. 
# a matlab style and an object oriented style.
# the matlab style is more likely to use plt.plot, 
# while an object oriented way is usually ax.plot
# where fig, ax = plt.subplots(2), for example

# ax will be an array of two Axes objects
fig, ax = plt.subplots(nrows=2, ncols=1)

# Call plot() method on the appropriate object
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x))
ax[0].plot(x, np.cos(x))
ax[1].plot(x, np.tan(x));
```
